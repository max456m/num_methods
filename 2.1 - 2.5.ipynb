```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return np.sin(x)
def f2(x):
  return 1/x

x1 = np.linspace(-20,20, 10000)
y1 = f1(x1)
x2 = np.linspace(-20,20,100)
y2 = f2(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0, color='k', linestyle='-')
plt.grid(True)
plt.title('График f(x)=cos(x) и f(x)=-3x/40')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
